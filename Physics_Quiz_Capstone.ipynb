```python
import random

quiz_questions = [
    {
        "q": "Which equation gives average velocity?",
        "opts": ["v = dx/dt", "v = at", "v = dx/t^2", "v = 2dx/t"],
        "ans": 0
    },
    {
        "q": "What is the formula for displacement starting from rest?",
        "opts": ["x = vt", "x = (1/2)*a*t^2", "x = at", "x = v^2/2a"],
        "ans": 1
    },
    {
        "q": "Newton's second law is?",
        "opts": ["F = mv", "F = ma", "F = m/a", "F = dp/dx"],
        "ans": 1
    },
    {
        "q": "Gravitational potential energy formula?",
        "opts": ["U = (1/2)*m*v^2", "U = mgh", "U = kx", "U = mgt"],
        "ans": 1
    },
    {    "q": "Kinetic energy formula?",
        "opts": ["KE = mgh", "KE = (1/2)*k*x^2", "KE = (1/2)*m*v^2", "KE = mv" ],
         "ans": 2
    },
   {    "q": "Momentum is defined as?",
        "opts": ["p = ma", "p = mv", "p = fv", "p = mv^2" ],
         "ans": 1
    },
   {    "q": "Work-energy theorem says work equals?",
        "opts": ["change in momentum", "change in kinetic energy", "force * time", "change in potential energy" ],
         "ans": 1
    },
   {    "q": "Hooke's Law for spring force?",
        "opts": ["F = -kx", "F = (1/2)*k*x^2", "F = mg", "F = kx^2" ],
         "ans": 0
    },
   {    "q": "Period of a simple pendulum?",
        "opts": ["T = 2pi*sqrt(m/k)", "T = 2pi*sqrt(L/g)", "T = 2pi*f", "T = 1/f" ],
         "ans": 1
    },
   {    "q": "Centripetal acceleration formula?",
        "opts": ["a = v/r", "a = v^2/r", "a = r*w", "a = v^2*r"],
         "ans": 1
    }
 ]



def run_quiz(questions):

    """Runs the quiz, presenting the title card, then presents the questions """


    print("Welcome to the Physics 1 Formula Quiz!")
    print()
    score = 0
    shuffled = questions.copy()
    random.shuffle(shuffled)

    letters = ["A", "B", "C", "D"]

    for item in shuffled:
        print(item["q"])
        for letter, opt in zip(letters, item["opts"]): # putting the list of quiz_quesitons & letters side by side - zip()
            print(f"{letter}: {opt}")

        correct_letter = letters[item["ans"]]
        answer = input("Choose the correct option: ").strip().upper()

        if answer == correct_letter:
            print()
            print("Correct!")
            score += 1
        else:
            print("Nope, better luck next time")
            print()
            print(f"The answer was {correct_letter}: {item['opts'][item['ans']]}")
        print()
        print(f"-"*40)

    print(f"Quiz Complete. Your final score is {score} out of {len(questions)}.")
    return score

def run(): # what happens when you run code
    while True:
        print()
        run_quiz(quiz_questions)
        again = input("Play again? (yes/no): ").strip().lower()
        if again != "yes":
            print("Thanks for playing!")
            return


run()
```
